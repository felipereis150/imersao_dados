### 1 - Realizar a média da metragem para cara um dos bairros (Paulo)







```python
media_metragem = dados.groupby('Bairro').mean('Metragem')
top_10_metragem = media_metragem.sort_values('Metragem', ascending = False)[:10].drop(columns =["Quartos", "Banheiros", "Vagas"])
top_10_metragem.plot.bar()
```
